```python
Content = pd.read_csv("yelp_restaurant_train.csv")

Content.drop(["review_id", "city","categories","avg_user_star","attributes","business_id","avg_business_stars","stars"], axis = 1, inplace = True)

dataframe_review = pd.DataFrame(Content)

dataframe_review.head(5)
```

```python
TempTable= pd.read_csv("yelp_restaurant_train.csv")
TempTable.drop(["review_id", "city","categories","avg_user_star","attributes","text"], axis = 1, inplace = True)
```

```python
TempTable['Label'] = TempTable.apply(lambda row: row.stars, axis = 1) 


TempTable.loc[TempTable['Label'] < 3, 'Label'] = -1
TempTable.loc[TempTable['Label'] > 3, 'Label'] = 1
TempTable.loc[TempTable['Label'] == 3, 'Label'] = 1
TempTable.Label = TempTable.Label.astype(int)
dfTempTable = pd.DataFrame(TempTable)

dfTempTable.head(5)
```

```python
merger = pd.concat([dataframe_review,dfTempTable],axis=1,join="inner")#, join_axes=[df_reviewGraph.user_id])
merger = merger.dropna()

newDS = merger[["Label","text"]] # add date later
newDS.head()

```

```python
newDS1 = pd.concat([newDS[newDS["Label"]==-1],newDS],axis=0)
newDS2 = pd.concat([newDS[newDS["Label"]==-1],newDS1],axis=0)
newDS3 = pd.concat([newDS[newDS["Label"]==-1],newDS2],axis=0)


OverSampledDS = pd.concat([newDS[newDS["Label"]==-1],newDS3],axis=0) 
```

```python
texts_Consider = []
labels_Consider = []

for i in range(len(OverSampledDS)):
    texts_Consider.append(OverSampledDS.iloc[i,1])
    labels_Consider.append(OverSampledDS.iloc[i,0])
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2), min_df=3)
classifier = LinearSVC()
Fit = vectorizer.fit_transform(texts_Consider)
  
cross_val_score = cross_val_score(classifier, Fit,labels_Consider, cv=2, n_jobs=-1)

print(cross_val_score)
```

```python
X_train, X_test, y_train, y_test = train_test_split(Fit, np.array(labels_Consider), test_size=0.4)
```

```python
classifier.fit(X_train, y_train)

p = classifier.predict(X_test)
```

```python
print("SVM Analyze")
print(classification_report(y_test, p))
```

```python
print("Confusion matrix ")
confusion_matrix(y_test,p)
```
